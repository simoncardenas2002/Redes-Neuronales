```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

```python
im= io.imread("../input/cnn-data-sources/Mara.jpeg")
```

```python
#es una imagen de 800 * 600 , de 3 capas
im.shape
```

```python
plt.imshow(im)
```

Separaremos las 3 capas de la imagen

```python
r= im[:,:,0]
g= im[:,:,1]
b= im[:,:,2]
```

```python
plt.imshow(g,cmap="gray")
```

Podemos trabajar esto como simplemente un arreglo

```python
plt.imshow(g.T,cmap="gray")
```

```python
r.shape
aux_dim=np.zeros([800,600])
```

```python
red=(np.dstack((r,aux_dim,aux_dim)).astype(np.uint8))
green=(np.dstack((aux_dim,g,aux_dim)).astype(np.uint8))
blue=(np.dstack((aux_dim,aux_dim,b)).astype(np.uint8))
```

```python
(np.dstack((r,aux_dim,aux_dim)).astype(np.uint8)).shape
```

 Al usar esa función dstack creamos 2 capas de 800 x 600 igualmente pero en este caso solo de ceros, esto es para poder ver la capa roja en este caso, por tanto si ponemos en cero los valores de las otras capas veremos de forma "pura" la respectiva capa, asi: 

```python
plt.imshow(red)
```

```python
plt.imshow(green)
```
