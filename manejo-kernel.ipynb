```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as nd
from skimage import io,color
```

Primero creamos la imagen con la que vamos a trabajar, recordar que siempre que podamos es mejor trabajar en una escala de grises

```python
img = io.imread("/kaggle/input/cnn-data-sources/yo.png")
img_gray=color.rgb2gray(img)
```

```python
plt.imshow(img)
print(img.shape)
```

```python
plt.imshow(img_gray)
print(img_gray.shape)
```

Vemos que al hacer el cambio a escala de grises pasamos a tener solo una capa, a diferencia de la imagen RGB donde teníamos 3 capas

```python
kernel= np.array([[-1,0,1],
                  [-1,0,1],
                 [-1,0,1]])
```

Este es el kernel con el cual puedo detectar bordes, este arreglo de 3x3 se utilizará para posteriormente hacer una convolución , que nos dará una nueva imagen

```python
#Hacemos la convolución con ayuda de la librería scipy
img_new= nd.convolve(img_gray,kernel)
```

```python
fig,axes = plt.subplots(1,2, figsize=(15,10))
axes[0].imshow(img_gray,cmap=plt.cm.gray)
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(img_new,cmap=plt.cm.gray)
axes[1].set_title("Convolution")
axes[1].axis("off")
```

Efectivamente vemos que este kernel nos ayuda a detectar exclusivamente los bordes "verticales"

Creemos otro kernel

```python
kernel2= np.array([[0,0,0],
                  [0,1,0],
                 [0,0,0]])
#Este kernel me da la misma imagen

img_new2= nd.convolve(img_gray,kernel2)

fig,axes = plt.subplots(1,2, figsize=(15,10))
axes[0].imshow(img_gray,cmap=plt.cm.gray)
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(img_new2,cmap=plt.cm.gray)
axes[1].set_title("Convolution")
axes[1].axis("off")
```

probaremos otros kernel, entender esto nos permite hallar distintas formas en las que podemos detectar características específicas en imágenes

```python
kernel3= np.array([[-1,0,-1],
                  [-1,0,-1],
                 [-1,0,-1]])
#Este kernel me da la misma imagen

img_new3= nd.convolve(img_gray,kernel3)

fig,axes = plt.subplots(1,2, figsize=(15,10))
axes[0].imshow(img_gray,cmap=plt.cm.gray)
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(img_new3,cmap=plt.cm.gray)
axes[1].set_title("Convolution")
axes[1].axis("off")
```
