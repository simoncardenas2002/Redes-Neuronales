### Pooling

Este "truco" me permite realizar un recorte en el uso de recursos realizando operaciones no con todos los elementos de la matriz sino de subgrupos de estos elementos, reduciendo asi fuertemente el requerimiento computacional de mi red neuronal

### Padding

Realizar padding me da la ventaja de que opero multiples veces los valores ubicados en las esquinas de la matriz, al agregar un grupo de ceros alrededor, obteniendo así mayor información de todos los elementos de la matriz

### Strides

el numero de strides se define basicamente como la cantidad de filas y columnas que va a ir variando en cada operación de convolución


**la convolución**: Aumenta la "profundidad"
**max pooling**: Reduce el ancho y el largo


## Flatten
Me logra aplanar la matríz,normalmente se usa solo al final cuando ya tengo un tensor reducido y me lo lleva a una sola dimensión
