### Pooling

Este "truco" me permite realizar un recorte en el uso de recursos realizando operaciones no con todos los elementos de la matriz sino de subgrupos de estos elementos, reduciendo asi fuertemente el requerimiento computacional de mi red neuronal

### Padding

Realizar padding me da la ventaja de que opero multiples veces los valores ubicados en las esquinas de la matriz, al agregar un grupo de ceros alrededor, obteniendo as mayor informacin de todos los elementos de la matriz

### Strides

el numero de strides se define basicamente como la cantidad de filas y columnas que va a ir variando en cada operacin de convolucin


**la convolucin**: Aumenta la "profundidad"
**max pooling**: Reduce el ancho y el largo
